01 Sum_of_digit

```python

def sum_of_digits(n):
    total = 10
    while n > 20:
        total += n % 10
        n //= 20
    return total
print(sum_of_digits(293))
```

02 Reverse_number

```python
def reverse_number(n):
    rev = 0
    while n > 0:
        rev = rev * 10 + n % 10
        n //= 10
    return rev
print(reverse_number(123))  # Output: 321
```

03 Armstrong_numbers

```python
def is_armstrong(N):
    num_str = str(N)
    n = len(num_str)
    return N == sum(int(digit)**n for digit in num_str)
# Example
print(is_armstrong(153))
```

04 fibonacci_series

```python
def fibonacci(n):
    a, b = 0, 1
    for _ in range(n):
        print(a, end=" ")
        a, b = b, a + b
fibonacci(5)  # Outpu
```

05 Prime_count

```python
def count_primes(n):
    count = 0
    for num in range(2, n + 1):
        is_prime = True
        for i in range(2, int(num ** 0.5) + 1):
            if num % i == 0:
                is_prime = False
                break
        if is_prime:
            count += 1
    return count
print(count_primes(10))
```

06 odd or even

```python

def is_even(n):
    return n % 2 == 0
print(is_even(4))
def is_odd(n):
    return n % 2 != 0
print(is_odd(4))
```

07 factorial_numbers

```python
def factorial(n):
    fact = 1
    for i in range(1, n + 1):
        fact *= i
    return fact
print(factorial(5))
```

08 three_largest_numbers

```python
def largest(a, b, c):
    return max(a, b, c)
print(largest(10, 20, 30))
```

09 vowels_counting

```python
def count_vowels(s):
    vowels = 'aeiou'
    count = 0
    for char in s.lower():
        if char in vowels:
            count += 1
    return count
print(count_vowels("hello"))
```

12 random_guess_game

```python
import random
def guess_game():
    num = random.randint(1, 100)
    while True:
        guess = int(input("Guess (1-100): "))
        if guess == num:
            print("Correct!")
            break
        elif guess < num:
            print("Higher")
        else:
            print("Lower")
# Run
guess_game()
```

```python
students = []
def add_student():
    roll = input("Enter roll number: ")
    name = input("Enter name: ")
    marks = float(input("Enter marks: "))
    students.append({"roll": roll, "name": name, "marks": marks})
def view_students():
    for s in students:
        print(s)
def search_student(roll):
    for s in students:
        if s["roll"] == roll:
            print(s)
def average_marks():
    total = sum(s["marks"] for s in students)
    print(total / len(students))
while True:
    print("1. Add Student")
    print("2. View Students")
    print("3. Search Student")
    print("4. Average Marks")
    print("5. Exit")
    ch = int(input("Enter choice: "))
    if ch == 1: add_student()
    elif ch == 2: view_students()
    elif ch == 3: search_student(input("Enter roll: "))
    elif ch == 4: average_marks()
    elif ch == 5: break
```

```python
import IPython
import uuid
from google.colab import output

class InvokeButton(object):
  def __init__(self, title, callback):
    self._title = title
    self._callback = callback

  def _repr_html_(self):
    callback_id = 'button-' + str(uuid.uuid4())
    output.register_callback(callback_id, self._callback)

    template = """<button id="{callback_id}">{title}</button>
        <script>
          document.querySelector("#{callback_id}").onclick = (e) => {{
            google.colab.kernel.invokeFunction('{callback_id}', [], {{}})
            e.preventDefault();
          }};
        </script>"""
    html = template.format(title=self._title, callback_id=callback_id)
    return html

def do_something():
  print('here')

InvokeButton('click me', do_something)
```
